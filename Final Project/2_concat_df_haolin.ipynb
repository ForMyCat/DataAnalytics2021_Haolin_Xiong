```python
import numpy as np
import pandas as pd
from netCDF4 import Dataset
import netCDF4
import netcdftime
import datetime

def nc4_to_df(meta_data):
  '''
  convert a single nc4 file to pandas df including four columns
  time, longitude, latitude, and precipitation 
  '''
  # meta_data = '/content/data/3B-DAY.MS.MRG.3IMERG.20160422-S000000-E235959.V06.nc4'

  nc_fid = netCDF4.Dataset(meta_data, "r", format="NETCDF4")


  time = nc_fid.variables['time'][:]
  t_unit = nc_fid.variables['time'].units
  t_cal = nc_fid.variables['time'].calendar

  nc4date = netcdftime.num2date(time,units = t_unit,calendar = t_cal)[0]
  date_time = datetime.datetime(nc4date.year,nc4date.month,nc4date.day)

  date_str = date_time.strftime("%m/%d/%Y")
  print("date and time:",date_str)

  lats = nc_fid.variables['lat'][:]  # extract/copy the data
  lons = nc_fid.variables['lon'][:]
  time = nc_fid.variables['time']
  t_unit = nc_fid.variables['time'].units

  precip = nc_fid.variables['precipitationCal'][:]

  latitudes_grid, longitudes_grid = [
      x.flatten() for x in np.meshgrid(lats, lons, indexing='ij')]
  df = pd.DataFrame({
      'latitude': latitudes_grid,
      'longitude': longitudes_grid,
      'precip': precip[:].flatten()})
  length = len(df)
  time_seris = pd.Series([date_str]*length)
  df['time'] = time_seris

  return df
# data_dict = nc_fid.variables
# for key in list(data_dict.keys()):
#   print(key, '----->', nc_fid.variables[key])
```

```python
'''
Query data with lat/lon range
'''
# Mali coordinate info
# lat_min, lat_max, lon_min, lon_max = 11.08943, 18.44111,  -11.44448, 3.14111

# Pakistan coordinate info
lat_min, lat_max, lon_min, lon_max = 24.3539, 35.91869,  61.74681, 75.16683
df = nc4_to_df('/content/drive/MyDrive/6962 final project/data/Daily_Precipitation/pakistan/3B-DAY.MS.MRG.3IMERG.20000911-S000000-E235959.V06.nc4')
```

```python
df['in_range'] = df.apply(lambda x: x['latitude']>=lat_min and x['longitude']>=lon_min and x['latitude']<=lat_max and x['longitude']<=lon_max,axis=1)
df_inrange = df.loc[df['in_range']]
range_index = df_inrange.index
```
