```python
def generate_urls(country = 'Pakistan'):
  '''
  Given a country name, generate a list of urls needed
  according to the alert dates of the country
  Retrieve previous 14 days' precipitation data of all alerts 
  '''
  df_loc = df.loc[df.groupby('country').groups[country]]
  print(df_loc)
  df_loc.reset_index(drop=True,inplace=True)
  df_loc['date'] = df_loc['issue_date'].apply(lambda x: datetime(x.year,x.month,x.day))
  df_loc = df_loc.loc[(df_loc['date'] >= datetime(2000, 5, 31))]

  date_list = list(set(df_loc['date']))
  date_list_2 = []
  for i in date_list:
    for j in range(0,15):
      date_list_2.append(i - timedelta(days = j))
  date_list_2 = list(set(date_list_2))
  date_list_2.sort()
  url_list = []
  for x in date_list_2:
    url_list.append(('https://gpm1.gesdisc.eosdis.nasa.gov/data/GPM_L3/GPM_3IMERGDF.06/'+str(x.year)+'/'+'{:02d}/3B-DAY.MS.MRG.3IMERG.'+ x.strftime("%Y%m%d") +'-S000000-E235959.V06.nc4').format(x.month))
  np.savetxt(country+"_url.txt",np.array(url_list), fmt = "%s")
```

```python
generate_urls(country = 'Pakistan')
```
